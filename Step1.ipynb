```python
def limpia(a):
    if '+' in str(a):
        return float(str(a).split('+')[0])
    elif '-' in str(a):
        return float(str(a).split('-')[0])
    else:
        return float(a)
```

```python
def elimina_tildes(cadena):
    s = ''.join((c for c in unicodedata.normalize('NFD',cadena) if unicodedata.category(c) != 'Mn'))
    return s
```

```python
def super_limpia(a):
    try:
        return int(a)
    except:
        return np.nan
```
