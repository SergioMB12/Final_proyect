```python
df_team.dropna(inplace=True) #elimino valores nulos

#df_team.drop_duplicates(inplace = True)
```

```python
def win(df_match):
    """
    Input: DataFrame home_team_goal, away_team_goal, home_team_name, away_team_name.Un total de 4 columnas
    OutPut: Columna con equipo ganador
    """
    home_s = df_match[0] #goles local  
    away_s = df_match[1] #goles visitante
    home_t = df_match[2] #nombre equipo local 
    away_t= df_match[3] #nombre equipo visitante
    
    if home_s > away_s:
        return home_t
    elif home_s < away_s:
        return away_t
    else:
        return 'DRAW'
# Aplico mi funcion y obtengo la columna
df_match['winner']  = df_match[['home_team_goal', 'away_team_goal', 'home_team_name', 'away_team_name']].apply(win, axis=1)
```

```python
def lose(df_match):
    
    home_score = df_match[0] 
    away_score = df_match[1] 
    home_team_name = df_match[2] 
    away_team_name = df_match[3] 
    
    if home_score < away_score:
        return home_team_name
    elif home_score > away_score:
        return away_team_name
    else:
        return 'DRAW'
    
df_match['loser']  = df_match[['home_team_goal', 'away_team_goal', 'home_team_name', 'away_team_name']].apply(lose, axis=1)
```
